```python
import requests
from bs4 import BeautifulSoup

# Fix the headers with correct line continuation
HEADERS = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/90.0.4430.212 Safari/537.36',
    'Accept-Language': 'en-US, en;q=0.5'
}

# user define function 
# Scrape the data
def getdata(url):
    r = requests.get(url, headers=HEADERS)
    return r.text

def html_code(url):
    # pass the url into getdata function
    htmldata = getdata(url)
    soup = BeautifulSoup(htmldata, 'html.parser')
    return soup

def get_product_info(soup):
    # Extract product name
    product_name = soup.find('span', {'id': 'productTitle'})
    product_name = product_name.get_text().strip() if product_name else "N/A"
    
    # Extract user rating (stars)
    user_rating = soup.find('span', {'class': 'a-icon-alt'})
    user_rating = user_rating.get_text().strip() if user_rating else "N/A"
    
    return product_name, user_rating

def get_reviews(soup):
    reviews = []
    
    review_elements = soup.find_all("div", class_="a-expander-content reviewText review-text-content a-expander-partial-collapse-content")
    
    for review_element in review_elements:
        review_text = review_element.get_text().strip()
        
        # Extract review date
        review_date = review_element.find_previous('span', class_="a-size-base a-color-secondary review-date")
        review_date = review_date.get_text().strip() if review_date else "N/A"
        
        reviews.append({
            'Review': review_text,
            'Review Date': review_date
        })
    
    return reviews

# Function to combine product info and reviews into a table
def get_product_and_reviews(url):
    soup = html_code(url)
    
    # Get product info (name and rating)
    product_name, user_rating = get_product_info(soup)
    
    # Get reviews
    reviews = get_reviews(soup)
    
    # Create table with product info and reviews
    result_table = []
    
    for review in reviews:
        result_table.append({
            'Product Name': product_name,
            'User Rating': user_rating,
            'Review Date': review['Review Date'],
            'Review': review['Review']
        })
    
    return result_table

# URL of the product (You can replace this with any Amazon product URL)
url = "https://a.co/d/d9ewY3S"
table = get_product_and_reviews(url)

# Display the table
import pandas as pd
df = pd.DataFrame(table)
df.head(100)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time

# Fix the headers with correct line continuation
HEADERS = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/90.0.4430.212 Safari/537.36',
    'Accept-Language': 'en-US, en;q=0.5'
}

# Function to scrape data from a given URL
def getdata(url):
    r = requests.get(url, headers=HEADERS)
    return r.text

# Function to parse the HTML content
def html_code(url):
    htmldata = getdata(url)
    soup = BeautifulSoup(htmldata, 'html.parser')
    return soup

# Function to extract product name and user rating
def get_product_info(soup):
    product_name = soup.find('span', {'id': 'productTitle'})
    product_name = product_name.get_text().strip() if product_name else "N/A"
    
    user_rating = soup.find('span', {'class': 'a-icon-alt'})
    user_rating = user_rating.get_text().strip() if user_rating else "N/A"
    
    return product_name, user_rating

# Function to extract reviews from the page
def get_reviews(soup):
    reviews = []
    review_elements = soup.find_all("div", class_="a-expander-content reviewText review-text-content a-expander-partial-collapse-content")
    
    for review_element in review_elements:
        review_text = review_element.get_text().strip()
        
        # Extract review date
        review_date = review_element.find_previous('span', class_="a-size-base a-color-secondary review-date")
        review_date = review_date.get_text().strip() if review_date else "N/A"
        
        reviews.append({
            'Review': review_text,
            'Review Date': review_date
        })
    
    return reviews

# Function to get reviews from multiple pages
def get_reviews_from_multiple_pages(product_url, num_pages=5):
    reviews = []
    
    # Get the product name and rating from the first page
    soup = html_code(product_url)
    product_name, user_rating = get_product_info(soup)
    
    # Loop through multiple pages
    for page in range(1, num_pages + 1):
        print(f"Scraping page {page}...")
        
        # Construct the URL for the current page
        paginated_url = f"{product_url}?pageNumber={page}"
        soup = html_code(paginated_url)
        
        # Extract reviews from the current page
        page_reviews = get_reviews(soup)
        
        for review in page_reviews:
            reviews.append({
                'Product Name': product_name,
                'User Rating': user_rating,
                'Review Date': review['Review Date'],
                'Review': review['Review']
            })
        
        # Add a delay to avoid being blocked by Amazon
        time.sleep(2)
    
    return reviews

# URL of the product (replace with your desired Amazon product URL)
url = "https://a.co/d/7UAaJHL"  # Replace with your Amazon product reviews page URL

# Try scraping reviews from multiple pages
reviews_data = get_reviews_from_multiple_pages(url, num_pages=5)

# Convert the data to a pandas DataFrame and display
if reviews_data:
    df = pd.DataFrame(reviews_data)
    print(df)
else:
    print("No reviews found or page structure may have changed.")
```
